```python
import os
data_files = os.listdir("data/")
```

```python
import pandas as pd
import pickle
import numpy as np
```

```python
dataframe = pd.DataFrame()
for file in data_files:
    with open(f'data/{file}', 'rb') as f:
        unpickled_frame = pickle.load(f)
        dataframe = dataframe.append(unpickled_frame)
```
